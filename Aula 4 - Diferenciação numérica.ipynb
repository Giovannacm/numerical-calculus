```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def avancada(func, x, h = 0.1):
    return (func(x + h) - func(x)) / h

def atrasada(func, x, h = 0.1):
    return (func(x) - func(x - h)) / h

def centrada(func, x, h = 0.1):
    return (func(x + h) - func(x - h)) / (2*h)

def centrada_2(func, x, h = 0.1):
    return (func(x - h) - 2 * func(x) + func(x + h)) / (h*h)
```

```python
f = np.sin
df = np.cos
def d2f(x):
    return -np.sin(x)
```

```python
x = np.linspace(0, np.pi, 100)

h = 0.5

aprox_avan = avancada(f, x, h)
aprox_atra = atrasada(f, x, h)
aprox_cent = centrada(f, x, h)

fig, ax = plt.subplots(1, 3)
fig.set_figwidth(15)

ax[0].plot(x, aprox_avan, 'b')
ax[0].plot(x, df(x), 'k--')
ax[0].set_title('Avancada')

ax[1].plot(x, aprox_atra, 'b')
ax[1].plot(x, df(x), 'k--')
ax[1].set_title('Atrasada')

ax[2].plot(x, aprox_cent, 'b')
ax[2].plot(x, df(x), 'k--')
ax[2].set_title('Centrada')
```

```python
#Observando o comportamento, vemos que na avançada (soma o erro) e na atrasada (subtrai o erro), existe um comportamento parecido, por causa do erro
```

```python
h = 0.1
seg_der = centrada_2(f, x, h)

plt.plot(x, seg_der, 'b')
plt.plot(x, d2f(x), 'k--')
```
