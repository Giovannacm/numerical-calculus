## 1) Considere a função f(x) = sin( exp(2x) ) no intervalo x \in [0,1].
### a) Utilize as fórmulas avançada, atrasada e centrada para aproximar a derivada de primeira ordem desta função no intervalo dado utilizando h = 0.5, 0.1 e 0.01. Plote um gráfico comparando os 3 métodos com a solução exata para cada valor de h. 

```python
def avancada(func, x, h = 0.1):
    return (func(x + h) - func(x)) / h

def atrasada(func, x, h = 0.1):
    return (func(x) - func(x - h)) / h

def centrada(func, x, h = 0.1):
    return (func(x + h) - func(x - h)) / (2*h)
```

```python
def f(x):
    return np.sin(np.exp(2*x))
def df(x):
    return 2*(np.cos(np.exp(2*x)))*(np.exp(2*x))
```

```python
x = np.linspace(0, 1, 100)

fig, ax = plt.subplots(3, 3)
fig.set_figheight(15)
fig.set_figwidth(15)


ax[0][0].plot(x, avancada(f, x, 0.5), 'b')
ax[0][0].plot(x, df(x), 'k--')
ax[0][0].set_title('Avancada (h = 0.5)')

ax[0][1].plot(x, atrasada(f, x, 0.5), 'b')
ax[0][1].plot(x, df(x), 'k--')
ax[0][1].set_title('Atrasada (h = 0.5)')

ax[0][2].plot(x, centrada(f, x, 0.5), 'b')
ax[0][2].plot(x, df(x), 'k--')
ax[0][2].set_title('Centrada (h = 0.5)')


ax[1][0].plot(x, avancada(f, x, 0.1), 'b')
ax[1][0].plot(x, df(x), 'k--')
ax[1][0].set_title('Avancada (h = 0.1)')

ax[1][1].plot(x, atrasada(f, x, 0.1), 'b')
ax[1][1].plot(x, df(x), 'k--')
ax[1][1].set_title('Atrasada (h = 0.1)')

ax[1][2].plot(x, centrada(f, x, 0.1), 'b')
ax[1][2].plot(x, df(x), 'k--')
ax[1][2].set_title('Centrada (h = 0.1)')


ax[2][0].plot(x, avancada(f, x, 0.01), 'b')
ax[2][0].plot(x, df(x), 'k--')
ax[2][0].set_title('Avancada (h = 0.01)')

ax[2][1].plot(x, atrasada(f, x, 0.01), 'b')
ax[2][1].plot(x, df(x), 'k--')
ax[2][1].set_title('Atrasada (h = 0.01)')

ax[2][2].plot(x, centrada(f, x, 0.01), 'b')
ax[2][2].plot(x, df(x), 'k--')
ax[2][2].set_title('Centrada (h = 0.01)')
```

```python
# Quando diminuímos o valor de h, é possível verificar que a aproximação se torna mais precisa.
```

### b) Repita o item anterior para a derivada de segunda ordem utilizando a fórmula centrada.

```python
def centrada_2(func, x, h = 0.1):
    return (func(x - h) - 2 * func(x) + func(x + h)) / (h*h)
```

```python
def d2f(x):
    return 4*(np.exp(2*x))*(np.cos(np.exp(2*x))) - 4*(np.exp(4*x))*(np.sin(np.exp(2*x)))
```

```python
fig, ax = plt.subplots(1, 3)
fig.set_figwidth(15)


ax[0].plot(x, centrada_2(f, x, 0.5), 'b')
ax[0].plot(x, d2f(x), 'k--')
ax[0].set_title('Centrada (h = 0.5)')


ax[1].plot(x, centrada_2(f, x, 0.1), 'b')
ax[1].plot(x, d2f(x), 'k--')
ax[1].set_title('Centrada (h = 0.1)')


ax[2].plot(x, centrada_2(f, x, 0.01), 'b')
ax[2].plot(x, d2f(x), 'k--')
ax[2].set_title('Centrada (h = 0.01)')
```

### c) Aproxime a derivada da primeira ordem da função f(x) no ponto x = 0.5 utilizando h = 10^(-i), i = 1, 2, ..., 10, utilizando as fórmulas avançada, atrasada e centrada. Plote o erro absoluto [abs(exato - aprox)] cometido por cada método em relação a cada valor de h.
#### obs: utilize os seguintes comandos após os plots para melhorar a visualização
#### plt.yscale('log')
#### plt.xscale('log')
#### plt.grid() 

```python
i = np.linspace(1, 10, 10)
h = 10**(-i)
aprox_avan = avancada(f, 0.5, h)
aprox_atra = atrasada(f, 0.5, h)
aprox_cent = centrada(f, 0.5, h)
plt.plot(h, np.abs(df(0.5) - aprox_avan), 'ro-')
plt.plot(h, np.abs(df(0.5) - aprox_atra), 'bo-')
plt.plot(h, np.abs(df(0.5) - aprox_cent), 'go-')
plt.yscale('log')
plt.xscale('log')
plt.legend(['Avançada', 'Atrasada', 'Centrada'])
plt.grid()
```

```python
# Quando diminuímos o valor de h, é possível verificar que, em geral, a aproximação se torna mais precisa, uma vez que o erro vai diminuindo.
# Além disso, o método centrado, como já é de se esperar, é aquele que apresenta um erro menor.
# Para o método avançado e atrasado, o valor de i que gera um erro menor é 9 (h = 10^(-9)). 
# Para o método centrado, o valor de i que gera um erro menor é 6 (h = 10^(-6)).
```
